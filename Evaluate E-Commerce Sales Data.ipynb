# Evaluate E-Commerce Sales Data

## Objective

The objective of this analysis is to discover the best-selling products based on the sales data.

## Introduction

The dataset records all transactions of an online retail company from December 1, 2010 to December 9, 2011, covering the company's business data in different countries and regions around the world. The company primarily sells gifts for various occasions, including birthday gifts, wedding favors, Christmas gifts, etc. Its customer base mainly consists of wholesalers and individual consumers, with wholesalers making up a significant proportion.

Definition of each column as below,

InvoiceNo: Invoice number, 6 digits. It is the transactions unique identifier. If this number started with "c", it means the transaction was cancelled.  
StockCode: Product code. 5 digits, it is the products unique identifier.  
Description: Description of product.  
Quantity: Quantity of product in the transaction.  
InvoiceDate: Date & time of invoice.  
UnitPrice: Unit price in Pound (£).  
CustomerID: Customer ID. 5 digits. Customers unique identifier.  
Country: Country.

## Read The Data

```python
import numpy as np
import pandas as pd
```

```python
original_data = pd.read_csv("e_commerce.csv")
original_data.head()
```

## Evaluate & Clean The Data

```python
cleaned_data = original_data.copy()
cleaned_data.head()
```

### Tidiness Of The Data

```python
cleaned_data.sample(10)
```

As seen from above, each variable as a column, each observation as a row, each cell contains a value, hence no issue.

### Cleanliness Of The Data

```python
cleaned_data.info()
```

Missing value exists in: Description & CustomerID

Dtype change to date: InvoiceDate  

Dtype change to str: CustomerID

#### Missing Value

```python
# Missing value in Description
```

```python
cleaned_data.query("Description.isna()")
```

There are 1454 rows of data with missing value in Description.  
From the output, all rows with missing value in Description have UnitPrice as 0.  
Hence, will further check if there is any rows with missing value in Description but not 0 in UnitPrice.

```python
cleaned_data.query("(Description.isna()) & (UnitPrice != 0)")
```

Both Description & UniPrice are critical variables for the analysis. The data without these 2 information will not be effective, hence should remove those obervations with missing value in Description. 

```python
cleaned_data = cleaned_data.dropna(subset=['Description'])
cleaned_data['Description'].isna().sum()
```

```python
# Missing value in CustomerID
```

```python
cleaned_data.query("CustomerID.isna()")
```

CustomerID is not an essential variable to analyze best-selling product. When the value in CustomerID is missing, the analysis won't be affected. Hence should keep the rows with missing values in CustomerID.

#### Data Type

```python
cleaned_data['InvoiceDate'] = pd.to_datetime(cleaned_data['InvoiceDate'])
cleaned_data['InvoiceDate']
```

```python
cleaned_data['CustomerID'] = cleaned_data['CustomerID'].astype(str)
cleaned_data['CustomerID']
```

```python
cleaned_data['CustomerID'] = cleaned_data['CustomerID'].str.slice(0,-2)
cleaned_data['CustomerID']
```

#### Duplicated Value

InvoiceNo, StockCode & CustomerID are unique identifier. However, one single transaction may include several products, same product may exist in different transaction & same customer may have different transaction. Hence, duplicated values are expected in InvoiceNo, StockCode & CustomerID.

#### Non-standardize Value

Check if non-standardize values exist in Country.

```python
cleaned_data['Country'].value_counts()
```

From above, United Kingdom, UK & U.K. refer to the same country. USA & United States refer to the same country.  
Hence, will replace UK & U.K. with United Kingdom, & replace USA with United States.

```python
cleaned_data['Country'] = cleaned_data['Country'].replace(['UK', 'U.K.'], 'United Kingdom')
```

```python
print(len(cleaned_data[cleaned_data['Country'] == 'UK']))
print(len(cleaned_data[cleaned_data['Country'] == 'U.K.']))
```

```python
cleaned_data['Country'] = cleaned_data['Country'].replace('USA', 'United States')
```

```python
print(len(cleaned_data[cleaned_data['Country'] == 'USA']))
```

#### Incorrect/Invalid Value

```python
cleaned_data.describe()
```

-ve value exists in Quantity & UnitPrice, this may affect subsequent analysis.

```python
# -ve value in Quantity
```

```python
cleaned_data.query('Quantity < 0')
```

From above, all observations with -ve value in Quantity have InvoiceNo start with C, meaning the transaction was cancelled.  
To further confirm, will check if there is any observation with -ve value in Quantiy but not InvoiceNo start with C.

```python
cleaned_data.query('(Quantity < 0) & (InvoiceNo.str[0] != "C")')
```

From above, there are observations with -ve value in Quantity but InvoiceNo not start with C.  
However, extracted observations all show UnitPrice as 0.  
Hence, will add in the criteria of UnitPrice to check.

```python
cleaned_data.query('(Quantity < 0) & (InvoiceNo.str[0] != "C") & (UnitPrice != 0)')
```

From above, when Quanity is -ve value, one of the below criterias is met,  
- InvoiceNo starts with C, meaning the transaction was cancelled
- UnitPrice as 0

This means these observations are invalid transaction & should be removed. 

```python
cleaned_data = cleaned_data[cleaned_data['Quantity'] >= 0]
len(cleaned_data[cleaned_data['Quantity'] < 0])
```

```python
# -ve value in UnitPrice
```

```python
cleaned_data.query('UnitPrice < 0')
```

From the output, observations with -ve value in UnitPrice are for adjust bad debt, not valid transaction.  
Hence, should remove these observations.

```python
cleaned_data = cleaned_data[cleaned_data['UnitPrice'] >= 0]
len(cleaned_data.query('UnitPrice < 0'))
```

### Sort The Data

```python
cleaned_data
```

Sort out the total quantity of each product sold in each country.

```python
product_quantity_groupby_country_and_stockcode = cleaned_data.groupby(['Country', 'StockCode'])['Quantity'].sum()
product_quantity_groupby_country_and_stockcode
```

```python
product_quantity_groupby_country_and_stockcode_df = product_quantity_groupby_country_and_stockcode.reset_index()
product_quantity_groupby_country_and_stockcode_df
```

Sort out the best-selling product in each country.

```python
country_best_selling_product = product_quantity_groupby_country_and_stockcode_df.groupby('Country')['Quantity'].max()
country_best_selling_product
```

Sort out the best-selling product in each country with stockcode.

```python
country_best_selling_product_with_stockcode = pd.merge(product_quantity_groupby_country_and_stockcode_df, country_best_selling_product, on=['Country', 'Quantity'])
country_best_selling_product_with_stockcode
```

List out all available products. 

```python
stockcode_with_description = cleaned_data[['StockCode', 'Description']].drop_duplicates()
stockcode_with_description
```

Sort out the best-selling product in each country with product description.

```python
country_best_selling_product_with_description = pd.merge(country_best_selling_product_with_stockcode, stockcode_with_description, on='StockCode')
country_best_selling_product_with_description
```

There are duplicated values in Description, hence will remove duplicates according to the unique identifier StockCode.

```python
country_best_selling_product_with_description = country_best_selling_product_with_description.drop_duplicates(subset=['Country', 'StockCode'])
country_best_selling_product_with_description
```

The best-selling product in each country are shown in above dataframe.
